```python
from pymongo import MongoClient
import pandas as pd
```

```python
#Conectando à base remota no Mongodb do AWS

mongo_client = MongoClient('18.229.104.131', 27017)
mongo_db = mongo_client['inatel']
mongo_collection = mongo_db['in242']
```

```python
# Criando o dataframe 

dados = list(mongo_collection.find())
df = pd.DataFrame(dados)
```

```python
# Aplicando tag de análise quente, frio e normal via função.

def check_temperatura(temperatura):
    if temperatura < 20:
        return 'Frio'
    if temperatura < 25:
        return 'Normal'
    return 'Quente'
```

```python
df['sensacao'] = df['temperatura'].apply(check_temperatura)
```

```python
#Conferindo df

df.head()
```

```python
#Contagem de classificação de temperatura - sensação

df.groupby(['sensacao']).count()
```

```python
# Gerar um DF consolidado com media, min e max e hora.
df2 = pd.DataFrame()
```

```python
df2 = df.set_index('data_coleta').resample('H')['temperatura'].agg(['mean','min','max'])
```

```python
df2
```

```python
# Removendo horas sem amostras do Dataframe
df3=df2.dropna(how='all')
```

```python
#conferindo o novo df criado
df3
```

```python
# Gráfico de barras da Amostra de df3
df3.plot.bar()
```

```python
# Criando a organização por média de temperatura descendente- maior para menor
df3_higher = df3.sort_values(by='mean', ascending=False)
```

```python
df3_higher
```

```python
# Eliminando dados máximo e mínimo
df3_higher.drop(['min', 'max'], axis=1, inplace=True)
```

```python
# Criando gráfico com os horários de maiores médias de temperatura
df3_higher.plot.bar()
```

```python
# Alterando atributos dos gráficos 

plt.style.available
plt.style.use('seaborn-dark-palette')
plt.rcParams['figure.figsize'] = (3,2)
```

```python
# Criando novo gráfico com cores e tamanho alterados
df3_higher.plot.bar()
```

```python
# Adicionando títulos aos gráficos
df3_higher.plot.bar()
plt.title('Media de temperatura') #adicionando o título
plt.xlabel('Hora')
plt.ylabel('Temperatura')
plt.show()
```
